```python
import pyUni10 as uni10
import sys
import numpy as np
import copy

def matSp():
  spin = 0.5
  dim = int(spin * 2 + 1)
  return uni10.Matrix(dim, dim, [0, 1, 0, 0]);

def matSm():
  spin = 0.5
  dim = int(spin * 2 + 1)
  return uni10.Matrix(dim, dim, [0, 0, 1, 0]);

def matSz():
  spin = 0.5
  dim = int(spin * 2 + 1)
  return uni10.Matrix(dim, dim, [0.5, 0, 0, -0.5]);

def Heisenberg():
        spin = 0.5
        sp = matSp();
        sm = matSm();
        sz = matSz();
        ham = uni10.otimes(sz, sz);
        ham += 0.5 * (uni10.otimes(sp, sm) + uni10.otimes(sm, sp));
        dim = int(spin * 2 + 1)
        bdi = uni10.Bond(uni10.BD_IN, dim);
        bdo = uni10.Bond(uni10.BD_OUT, dim);
        H = uni10.UniTensor([bdi, bdi, bdo, bdo], "Heisenberg");
        H.putBlock(ham)
        return H


def TFIM(h):
    spin = 0.5
    sx = 0.5*(matSp()+matSm())
    sz = matSz()
    iden = uni10.Matrix(2,2, [1, 0, 0, 1])
    ham =uni10.otimes(sz,sz) +0.5*float(h)*(uni10.otimes(iden,sx)+uni10.otimes(sx,iden))
    dim = int(spin * 2 + 1)
    bdi = uni10.Bond(uni10.BD_IN, dim);
    bdo = uni10.Bond(uni10.BD_OUT, dim);
    H =  uni10.UniTensor([bdi, bdi, bdo, bdo], "TFIM");
    H.putBlock(ham)
    return H





def bondcat(T, L, bidx):
    labels = T.label();
    per_labels = list(T.label())
    #print 'before:', per_labels
    per_labels.insert(0, per_labels.pop(bidx))
    #print 'after:', per_labels
    inBondNum = T.inBondNum();
    T.permute(per_labels, 1)
    T.putBlock(L * T.getBlock())
    T.permute(labels, inBondNum);
    return T

def bondrm(T, L, bidx):
    invL = uni10.Matrix(L.row(), L.col(), True)
    for i in xrange(L.elemNum()):
        invL[i] = 0 if L[i] == 0 else 1.0 / L[i]
    return bondcat(T, invL, bidx)

chi = 3
delta = 0.02
N = 10000
H = Heisenberg()
z=2

bdi_chi = uni10.Bond(uni10.BD_IN, chi);
bdo_chi = uni10.Bond(uni10.BD_OUT, chi);
bond_list=[]
for i in range(z):
    bond_list.append(bdi_chi)
bond_list.append(H.bond(2))

Gs = []
Gs.append(uni10.UniTensor(bond_list, "Ga"))
Gs.append(uni10.UniTensor(bond_list, "Gb"))
Gs[0].randomize(), Gs[1].randomize()


I_chi = uni10.Matrix(chi, chi, True)
I_chi.randomize()
Ls = []

for i in range(z):
    Ls.append(uni10.Matrix(chi, chi, True))  # Diagonal matrix
    Ls[i]=copy.copy(I_chi)
    
    

U = uni10.UniTensor(H.bond(), "U");
U.putBlock(uni10.takeExp(-delta, H.getBlock()))


def update(G, L, z, N):
    """
    fsdaf
    """
    for i in range(N):

        for bond in range(z):
            #print 'bond:', bond
            Gs_A_l=copy.copy(Gs[0])
        
            per_labels=list(Gs_A_l.label())
            per_labels.pop()
            per_labels.pop(bond)
        
            for b in range(z-1,0, -1):
               # print 'b:',b
                Gs_A_l=bondcat(Gs_A_l, Ls[(bond+b) % z], b)
            
            A_merged_bonds=per_labels
            #print "A Bonds to be merged:", A_merged_bonds
            if len(A_merged_bonds) > 1 :
                Gs_A_l.combineBond(A_merged_bonds)       
            #print 'A after combine:', Gs_A_l.label()
            Gs_A_l=bondcat(Gs_A_l,Ls[bond],0)
            Gs_A_l.setName("Gs_A_l")
        
            Gs_B_l=copy.copy(Gs[1])

            per_labels=list(Gs_B_l.label())
            per_labels.pop()
            per_labels.pop(bond)
            
            for b in range(z-1,0, -1): 
                Gs_B_l=bondcat(Gs_B_l, Ls[(bond+b) % z], b)
            B_merged_bonds=per_labels
            
            #print "B Bonds to be merged:",B_merged_bonds
            if len(B_merged_bonds) > 1:
                Gs_B_l.combineBond(B_merged_bonds)       
            #print 'B after combine:',Gs_B_l.label()
       
            Gs_A_l.setLabel([3, -1, 1]);
            Gs_B_l.setLabel([3, -3, 2]);
            U.setLabel([1, 2, -2, -4]);
            print Gs_A_l,Gs_B_l
            theta = uni10.contract(Gs_A_l, Gs_B_l, True) # Gs[A], Gs[B] is permuted atfer the execution
            theta *= U;
            theta.permute([-1, -2, -3, -4], 2);

            
        
            # SVD
            svd = theta.getBlock().svd()

            # Truncation
            sv = svd[1]
            norm = sv.resize(chi, chi).norm()
            sv = sv * (1.0 / norm);
            Ls[bond] = sv


            Gs_A_l.putBlock(svd[0].resize(svd[0].row(),chi));
            Gs_B_l.putBlock(svd[2].resize(chi,svd[2].col()));
            Gs_A_l.permute([3,-1,1],2)
            
            #Gs_A_l.setName("Gs_A_l"+str(i))
            Gs_B_l.permute([3,-3,2],2)
            #Gs_B_l.setName("Gs_AB_l"+str(i))
            
            print "After SVD\n",Gs_A_l,Gs_B_l

            labels=list(Gs[1].label())
            per_labels = list(Gs[1].label())

            per_labels.insert(0, per_labels.pop(bond))
            #print "A:", per_labels

            Gs[0].permute(per_labels,z)
            Gs[0].putBlock(Gs_A_l.getBlock())
        
            Gs[0].permute(labels,z)
        
            labels=list(Gs[1].label())
            per_labels = list(Gs[1].label())

            per_labels.insert(0, per_labels.pop(bond))
            #print "B:", per_labels
            Gs[1].permute(per_labels,z)
            Gs[1].putBlock(Gs_B_l.getBlock())
            Gs[1].permute(labels,z)
        
            for i in A_merged_bonds:
                Gs[0] = bondrm(Gs[0], Ls[i], i);
            for i in B_merged_bonds:
                Gs[1] = bondrm(Gs[1], Ls[i], i);

        val = (theta * theta)[0]
        print "E =", -np.log(val) / delta / 2
        

        

update(Gs,Ls,z, 2)



```
